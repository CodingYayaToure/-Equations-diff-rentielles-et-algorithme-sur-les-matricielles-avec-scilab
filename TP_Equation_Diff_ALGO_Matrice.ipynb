# Pour résoudre l'équation différentielle donnée $ y'' = -4y $ avec les conditions initiales $ y(0) = 3 $ et $ y'(0) = 0 $, nous pouvons transformer cette équation d'ordre 2 en un système de deux équations d'ordre 1.

### Étapes :

1. **Introduire de nouvelles variables :**

   Posons $y_1 = y$  et $y_2 = y'$.

2. **Réécrire l'équation d'ordre 2 :**

   L'équation $ y'' = -4y $ devient $ y_2' = -4y_1 $.

   Nous avons aussi $ y_1' = y_2 $.

3. **Former le système d'équations d'ordre 1 :**

   Le système devient :
   $
   \begin{cases}
   y_1' = y_2 \\
   y_2' = -4y_1
   \end{cases}
   $

4. **Conditions initiales :**

   Les conditions initiales sont :
   $
   \begin{cases}
   y_1(0) = 3 \\
   y_2(0) = 0
   \end{cases}
   $


### Explications :

1. **Définir la fonction `odefun`** :
   - `dydt = [y(2); -4*y(1)]` représente le système d'équations différentielles.

2. **Conditions initiales** :
   - `y0 = [3; 0]` représente les conditions initiales $ y_1(0) = 3 $ et $ y_2(0) = 0 $.

3. **Intervalle de temps** :
   - `t = linspace(0, 10, 100)` définit un intervalle de temps de 0 à 10 avec 100 points.

4. **Résoudre le système** :
   - `y = ode(y0, t, odefun)` utilise la fonction `ode` pour résoudre le système d'équations différentielles.

5. **Tracer les solutions** :
   - `plot(t, y(1, :), 'r-', t, y(2, :), 'b--')` trace les solutions $ y_1 $ et $ y_2 $ sur le même graphique.

6. **Légendes et titres** :
   - `legend("$y_1 = y$", "$y_2 = y'$")` ajoute des légendes aux courbes en utilisant des notations mathématiques.
   - `xlabel`, `ylabel`, et `title` ajoutent des légendes et des titres appropriés au graphique.



```scilab
// Utiliser funcprot pour éviter la redéfinition de la fonction
funcprot(0);

// Définir la fonction qui représente le système d'équations
function dydt = system(t, y)
    dydt = zeros(2, 1); // Initialiser le vecteur de sortie
    dydt(1) = y(2);     // y' = v
    dydt(2) = -4 * y(1); // v' = -4y
endfunction

// Conditions initiales
y0 = [3; 0]; // y(0) = 3, y'(0) = 0

// Intervalle de temps pour la solution
t0 = 0; // Temps initial
tf = 10; // Temps final
t = linspace(t0, tf, 100); // Créer un vecteur de temps

// Résoudre le système d'équations avec ode
y = ode(y0, t0, t, system);

// Tracer les résultats
clf; // Effacer la figure précédente
plot(t, y(1, :), 'r', 'LineWidth', 2); // Tracer y(t)
plot(t, y(1, :), 'r-', t, y(2, :), 'b--'); // Tracer y1 et y2 sur le même graphique

xlabel('Temps t');
ylabel('y(t)');
title('Solution de l''équation différentielle y'''' = -4y'); // Échapper l'apostrophe
// Légendes et titres
legend("y_1 = y", "y_2 = y''"); // Ajouter des légendes aux courbes
xgrid; // Activer la grille
```

```scilab
// Définir la fonction odefun
function dydt = odefun(t, y)
    dydt = [y(2); -4 * y(1)]; // Système d'équations différentielles
endfunction

// Conditions initiales
y0 = [3; 0]; // y(0) = 3 et y'(0) = 0

// Intervalle de temps
t = linspace(0, 10, 100); // Créer un vecteur de temps de 0 à 10 avec 100 points

// Résoudre le système
y = ode(y0, 0, t, odefun); // Utiliser la fonction ode pour résoudre le système d'équations

// Tracer les solutions
clf; // Effacer la figure précédente
plot(t, y(1, :), 'r-', t, y(2, :), 'b--', 'LineWidth', 4); // Tracer y1 et y2 sur le même graphique

// Légendes et titres
legend("y_1 = y", "y_2 = y''"); // Ajouter des légendes aux courbes
xlabel('Temps (t)'); // Étiqueter l'axe des x
ylabel('y(t)'); // Étiqueter l'axe des y
title("Solutions de l''équation différentielle y'' = -4y"); // Titre du graphique
xgrid; // Activer la grille
```

# EXEMPLE 2


# Pour résoudre l'équation différentielle de Van der Pol donnée par $y'' = c(1 - y^2)y' - y$ en utilisant Scilab, nous pouvons la reformuler en un système de deux équations différentielles du premier ordre.

### Étapes :

1. **Introduire de nouvelles variables :**
   Posons  $u_1(t) = y(t)$  et $ u_2(t) = y'(t)$.

2. **Réécrire l'équation d'ordre 2 :**
   L'équation $y'' = c(1 - y^2)y' - y $ devient $ u_2' = c(1 - u_1^2)u_2 - u_1$.

   Nous avons aussi $ u_1' = u_2$.

3. **Former le système d'équations d'ordre 1 :**
   Le système devient :
   $
   \begin{cases}
   u_1' = u_2 \\
   u_2' = c(1 - u_1^2)u_2 - u_1
   \end{cases}
   $


### Explications des modifications :
1. **Définition de `odefun` :** La fonction `odefun` prend maintenant le paramètre \( c \) et définit le système d'équations différentielles.
2. **Conditions initiales :** J'ai utilisé des conditions initiales \( y(0) = 2 \) et \( y'(0) = 0 \) comme exemple. Vous pouvez les modifier selon vos besoins.
3. **Intervalle de temps :** Le vecteur de temps est défini de 0 à 20 avec 1000 points pour une meilleure résolution.
4. **Résolution du système :** Utilisation de la fonction `ode` avec les conditions initiales, l'intervalle de temps et le paramètre \( c \).
5. **Tracé des solutions :** Les solutions $( u_1 )$ et $( u_2 )$ sont tracées sur le même graphique avec des légendes appropriées.


```scilab
// Définir la fonction odefun pour le système de Van der Pol
function dydt = odefun(t, y)
    c = 0.1; // Paramètre c
    dydt = [y(2); c * (1 - y(1)^2) * y(2) - y(1)]; // Système d'équations différentielles
endfunction

// Conditions initiales
y0 = [2; 0]; // y(0) = 2 et y'(0) = 0

// Intervalle de temps
t = linspace(0, 20, 1000); // Créer un vecteur de temps de 0 à 20 avec 1000 points

// Résoudre le système
y = ode(y0, 0, t, odefun); // Utiliser la fonction ode pour résoudre le système d'équations

// Tracer les solutions
clf; // Effacer la figure précédente
plot(t, y(1, :), 'r-', t, y(2, :), 'b--', 'LineWidth', 4); // Tracer y1 et y2 sur le même graphique

// Légendes et titres
legend('u_1 = y(t)', 'u_2 = y''(t)'); // Ajouter des légendes aux courbes
xlabel('Temps (t)'); // Étiqueter l'axe des x
ylabel('Valeur'); // Étiqueter l'axe des y
title('Solutions de l''équation de Van der Pol'); // Titre du graphique
xgrid; // Activer la grille
```
